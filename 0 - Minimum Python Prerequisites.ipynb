```python
# prompt: create a python script with an array with all the planets of the solar system and a random function that choses a random planet and print its name via console

import random

# Create a list of planets
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

# Define a function to choose a random planet
def choose_random_planet():
  # Generate a random index within the range of the planets list
  random_index = random.randint(0, len(planets) - 1)

  # Return the planet at the random index
  return planets[random_index]

# Choose a random planet and print its name
random_planet = choose_random_planet()
print(f"The chosen planet is: {random_planet}")
```

Exercise 1: Write a function make_more_extreme that has the following behavior:

    the function subtracts one from negative numbers
    the function adds one to positive numbers
    the function doesn't affect 0
    

```python
def make_more_extreme(x):
  if x<0:
    return x-1
  if x>0:
    return x+1
  else:
    return 0

print (make_more_extreme(1))
```
